# DATA ANALYSIS

```python
#mengkonversikan kolom waktu menjadi format datetime
df['tapInTime'] = pd.to_datetime(df['tapInTime'], errors='coerce')
df['tapOutTime'] = pd.to_datetime(df['tapOutTime'], errors='coerce')
df['journey_duration'] = (df['tapOutTime'] - df['tapInTime']).dt.total_seconds() / 60  # dalam menit
df = df[df['journey_duration'].notnull() & (df['journey_duration'] > 0)]
```

##Kalkulasi Metrik Efesiensi

## Rata-rata Durasi Perjalanan Per Koridor

```python
corridor_efficiency = df.groupby('corridorName')['journey_duration'].mean().reset_index()
corridor_efficiency.rename(columns={'journey_duration': 'average_journey_time'}, inplace=True)
corridor_efficiency
```

##Rata-rata Durasi Perjalan per Halte

```python
stop_efficiency = df.groupby(['tapInStopsName', 'tapOutStopsName'])['journey_duration'].mean().reset_index()
stop_efficiency.rename(columns={'journey_duration': 'average_stop_duration'}, inplace=True)
stop_efficiency
```

## Koridor dengan Rata-rata Durasi Pemberhentian di Atas Rata-rata

```python
avg_journey_time = corridor_efficiency['average_journey_time'].mean()
inefficient_corridors = corridor_efficiency[corridor_efficiency['average_journey_time'] > avg_journey_time]
avg_journey_time
```

## Halte dengan Rata-rata Durasi Pemberhentian di Atas Rata-rata:

```python
avg_stop_duration = stop_efficiency['average_stop_duration'].mean()
inefficient_stops = stop_efficiency[stop_efficiency['average_stop_duration'] > avg_stop_duration]
```
